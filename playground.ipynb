```python
from PIL import Image
import numpy as np
```

```python
# Open the image
raster_path = "./data/images/ee-satellite-zoom20.png"
# raster_path = get_data("OSBS_029.tif")
img = Image.open(raster_path)


# Ensure the image has 4 channels (RGBA)
if img.mode == 'RGBA':
    # Convert the image to RGB
    img = img.convert('RGB')
    # Create new path by adding "new_no_alpha_" at the beginning of the file name
    dir_name = os.path.dirname(raster_path)
    base_name = os.path.basename(raster_path)
    new_path = os.path.join(dir_name, "new_no_alpha_" + base_name)

    # Save the image without the alpha channel
    img.save(new_path)

    raster_path = new_path

# Window size of 300px with an overlap of 25% among windows for this small tile.
predicted_raster_w_p_800 = model.predict_tile(raster_path, return_plot = True, color=(0,165,255), patch_size=800,patch_overlap=0.25)
```
